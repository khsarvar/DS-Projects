# Import the library

```python
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv("test.csv")
test_Ids = test["PassengerId"]

train.head()
```

# Explatory Data Analysis

```python
print(test.shape)
print(train.shape)
```

```python
print(train.isna().sum())
print("-"*10)
print(test.isna().sum())
```

```python
def clean(data):
    data = data.drop(["Name", "Cabin", "Ticket", "PassengerId"], axis = 1)
    #data["Sex"] = data["Sex"].map({"male":1, "female":0})
    data["Fare"].fillna(data["Fare"].median(), inplace = True)
    data["Age"].fillna(data["Age"].median(), inplace = True)
    data["Embarked"].fillna("U", inplace = True)
    return data
```

```python
clean_train = clean(train)
clean_test = clean(test)
```

```python
clean_train.head(5)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

y = clean_train['Survived']
X = clean_train.drop("Survived", axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 2)
```

```python
clf = LogisticRegression(random_state = 2, max_iter=1000).fit(X_train, y_train)
```

```python
predictions = clf.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
submission_pred = clf.predict(clean_test)
df = pd.DataFrame({"PassengerId":test_Ids.values, "Survived":submission_pred})
df
```

```python
df.to_csv("submission.csv", index=False)
```
